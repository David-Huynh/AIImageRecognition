```python
def _bytes_feature(value):
    """Returns a bytes_list from a string / byte."""
    if isinstance(value, type(tf.constant(0))):
        value = value.numpy()
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))

def _float_feature(value):
    """Returns a float_list from a float / double."""
    return tf.train.Feature(float_list=tf.train.FloatList(value=value))

def serialize_example(image, label):
    """Convert image and label to tf.Example format."""
    # 이미지를 인코딩
    image = tf.io.encode_jpeg(image)
    
    feature = {
        'image': _bytes_feature(image),
        'label': _float_feature(label.numpy())
    }
    
    example_proto = tf.train.Example(features=tf.train.Features(feature=feature))
    return example_proto.SerializeToString()

def visualize_dataset_samples(dataset, title, num_samples=5):
    """데이터셋 샘플 시각화"""
    import matplotlib.pyplot as plt
    
    plt.figure(figsize=(15, 3))
    for i, (image, label) in enumerate(dataset.take(num_samples)):
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(image)
        plt.title(f'Label: {tf.argmax(label)}')
        plt.axis('off')
    plt.suptitle(title)
    plt.show()

def print_dataset_stats(dataset, name):
    """데이터셋 통계 출력"""
    total_samples = 0
    label_counts = {}
    
    for images, labels in dataset:
        batch_size = len(images)
        total_samples += batch_size
        
        # 레이블 카운트
        for label in labels:
            label_idx = tf.argmax(label).numpy()
            label_counts[label_idx] = label_counts.get(label_idx, 0) + 1
    
    print(f"\n{name} 데이터셋 통계:")
    print(f"총 샘플 수: {total_samples}")
    print("레이블 분포:")
    for label, count in label_counts.items():
        print(f"- 레이블 {label}: {count}개 ({count/total_samples*100:.2f}%)")

def save_to_tfrecord(dataset, filename):
    """데이터셋을 단일 TFRecord 파일로 저장"""
    print(f"\nSaving to {filename}...")
    
    # 저장하기 전에 데이터셋 통계 출력 및 샘플 시각화
    print_dataset_stats(dataset, os.path.basename(filename))
    visualize_dataset_samples(dataset.unbatch(), f"샘플 이미지 - {os.path.basename(filename)}")
    
    with tf.io.TFRecordWriter(filename) as writer:
        total_samples = 0
        for images, labels in dataset:
            for image, label in zip(images, labels):
                example = serialize_example(image, label)
                writer.write(example)
                total_samples += 1
                
                # 진행상황 표시
                if total_samples % 100 == 0:
                    print(f"처리된 샘플: {total_samples}")
    
    print(f"Saved {total_samples} samples to {filename}")

# 각 모델별로 데이터셋 생성 및 저장
for model_name, img_size in models.items():
    print(f"\nProcessing {model_name}...")
    
    # Training
    print("- Creating training dataset...")
    train_ds = create_basic_dataset(train_paths, train_labels, img_size, is_training=True)
    train_path = f"{save_dirs['train']}/{model_name}_train.tfrecord"
    save_to_tfrecord(train_ds, train_path)
    
    # Validation
    print("- Creating validation dataset...")
    val_ds = create_basic_dataset(val_paths, val_labels, img_size, is_training=False)
    val_path = f"{save_dirs['val']}/{model_name}_val.tfrecord"
    save_to_tfrecord(val_ds, val_path)
```

```python
import os
import torch
import tensorflow as tf
from tqdm import tqdm

def parse_tfrecord(example_proto):
    feature_description = {
        'image': tf.io.FixedLenFeature([], tf.string),
        'label': tf.io.FixedLenFeature([2], tf.float32)
    }
    example = tf.io.parse_single_example(example_proto, feature_description)
    image = tf.io.decode_jpeg(example['image'], channels=3)
    image = tf.image.resize(image, (256, 256))
    image = tf.cast(image, tf.float32) / 255.0
    return image, example['label']

def convert_tfrecord_to_individual_pt(tfrecord_path, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    dataset = tf.data.TFRecordDataset(tfrecord_path)
    dataset = dataset.map(parse_tfrecord)

    print(f"📦 변환 시작: {tfrecord_path} → {output_dir}")
    count = 0
    for image, label in tqdm(dataset):
        img = torch.tensor(image.numpy()).permute(2, 0, 1).contiguous()
        lbl = torch.tensor(label.numpy())

        torch.save({'image': img, 'label': lbl}, os.path.join(output_dir, f"{count:05d}.pt"))
        count += 1

    print(f"✅ 변환 완료. 총 {count}개 저장됨")

# 경로 지정
train_tfrecord = "records/train/swin_transformer_train.tfrecord"
val_tfrecord = "records/val/swin_transformer_val.tfrecord"

train_out_dir = "records/train_pt"
val_out_dir = "records/val_pt"

# 실행
convert_tfrecord_to_individual_pt(train_tfrecord, train_out_dir)
convert_tfrecord_to_individual_pt(val_tfrecord, val_out_dir)
```
